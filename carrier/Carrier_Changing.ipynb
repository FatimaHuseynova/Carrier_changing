```python

# Count the total number of workers
total_workers = len(data)

# Count the number of male workers
male_count = (data["Sex"] == "Male").sum()

# Calculate the percentage of male workers
percentage_male = (male_count / total_workers) * 100

print(f"Total workers: {total_workers}")
print(f"Male workers: {male_count}")
print(f"Percentage of male workers: {percentage_male:.2f}%")
```

```python
# Assuming you already have the total_workers and male_count calculated

# Count the number of female workers
female_count = total_workers - male_count

# Calculate the percentage of female workers
percentage_female = (female_count / total_workers) * 100

print(f"Total workers: {total_workers}")
print(f"Female workers: {female_count}")
print(f"Percentage of female workers: {percentage_female:.2f}%")
```

```python
import pandas as pd
import re
# Clean salary values and extract numeric part
data["CleanedSalary"] = data["Current salary"].apply(lambda x: float(re.sub(r"[^0-9.]", "", x)))

# Count unique salary types
unique_salary_types = data["CleanedSalary"].nunique()
```

```python
# Calculate the average salary by industry
average_salaries = data.groupby("Industry")["CleanedSalary"].mean()

top_industries = average_salaries.sort_values(ascending=False).head(5)

print("Top 5 industries with the highest average salary:")
print(top_industries)
```

```python
# Remove the prefix "Other (Please Specify):"
data["Industry"] = data["Industry"].str.replace("Other (Please Specify):", "")
```

```python

# Filter those who switched their career to data
switched_to_data = data[data["Switch carreer to data?"] == "Yes"]

# Count how many find working with data "easy"
easy_count = switched_to_data["Is it difficult to work with data?"].value_counts().get("Easy", 0)

# Calculate the percentage
total_switched_to_data = len(switched_to_data)
percentage_easy = (easy_count / total_switched_to_data) * 100

print(f"What percentage of people who change majors think that working with data is easy?: {percentage_easy:.2f}%")
```

```python
# Clean data: Remove rows with "Other (Please Specify):" in Current Role
data['Your current role'] = data['Your current role'].apply(lambda x: x.split(':')[1].strip() if ':' in x else x)

# Filter data for satisfaction > 8
high_satisfaction = data[data['Satisfaction with salary'] > 8]

# Count occurrences of each country
top_countries = high_satisfaction['Country'].value_counts().head(5)

# Plotting
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Satisfaction > 8')
plt.xlabel('Country')
plt.ylabel('Number of Workers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Output the top 5 countries and their workers' roles
top_country_names = top_countries.index.tolist()
for country in top_country_names:
    roles = high_satisfaction[high_satisfaction['Country'] == country]['Your current role'].unique()
    print(f"\nCountry: {country}")
    print("Workers' Roles:")
    for role in roles:
        print(role)
```

```python
data["Your current role"] = data["Your current role"].str.replace("Other (Please Specify):", "")
data["Country"] = data["Country"].str.replace("Other (Please Specify):", "")
```

```python
# Filter data for satisfaction score greater than 8
filtered_data = data[data["Satisfaction with salary"] > 8]
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Country", y="Your current role", hue="Satisfaction with salary", data=filtered_data, palette="coolwarm")
plt.title("Scatter Plot: Satisfaction vs. Current Role by Country")
plt.xlabel("Country")
plt.ylabel("Current Role")
plt.xticks(rotation=45)
plt.legend(title="Satisfaction")
plt.tight_layout()
plt.show()
```
